```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the stock tickers for the companies
companies = {
    'Sanofi India': 'SANOFI.NS',
    'GlaxoSmithKline': 'GLAXO.NS',
    'EIH Associated Hotel': 'EIHAHOTELS.NS',
    'Orient Hotel': 'ORIENTHOT.NS'
}

# Download historical data for the companies
historical_data = {}
for company, ticker in companies.items():
    historical_data[company] = yf.download(ticker, start='2020-01-01', end='2024-09-07')

# Plotting separate graphs for each company's volume with 180-day moving average and price (open, close, low, high)
for company, data in historical_data.items():
    if not data.empty:
        # Calculate 180-day moving average for volume
        data['Volume_MA_180'] = data['Volume'].rolling(window=180).mean()

        # Create a figure with two subplots
        fig, ax1 = plt.subplots(figsize=(14, 7))

        # Plot the price data
        ax1.set_title(company + ' Price and Volume Analysis')
        ax1.plot(data.index, data['Open'], label='Open', color='green')
        ax1.plot(data.index, data['Close'], label='Close', color='blue')
        ax1.plot(data.index, data['Low'], label='Low', color='red')
        ax1.plot(data.index, data['High'], label='High', color='orange')
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Price (INR)')
        ax1.legend(loc='upper left')

        # Create a second y-axis for the volume
        ax2 = ax1.twinx()
        ax2.plot(data.index, data['Volume'], label='Volume', color='gray', alpha=0.3)
        ax2.plot(data.index, data['Volume_MA_180'], label='180-Day MA Volume', color='purple')
        ax2.set_ylabel('Volume')
        ax2.legend(loc='upper right')

        plt.grid(True)
        plt.show()
```
